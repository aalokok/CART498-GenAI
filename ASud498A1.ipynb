```python
from diffusers import StableDiffusion3Pipeline

pipe = StableDiffusion3Pipeline.from_pretrained("stabilityai/stable-diffusion-3.5-medium", torch_dtype=torch.bfloat16)
pipe = pipe.to("cuda")
```

```python
image = pipe(
    "a composition depicting decay of human identity through time, in surrealistic art style inpsired by max ernst , dadaism style illumination, composition in style of cut-out collage, use cold tones  ",
    num_inference_steps=40,
    guidance_scale=2,
).images[0]
image.save("surreal_collage_2.png")
```
